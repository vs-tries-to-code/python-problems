5. Generate all permutations of a string, recursively

---



```python
def permute(s, left, right):
  if left==right:
    return (s)
  else:
    for i in range(left, right+1):
      l=list(s)
      l[left],l[i]=l[i],l[left]
      return permute("".join(s), left+1, right)

s=input("Enter string: ")
print(permute(s,0,len(s)-1))
```

# **Dynamic Programming**

1. Climbing stars having n steps

```python
n=int(input("Enter number of steps in stairs: "))
dp=(n+1)*[0]
dp[0]=0
dp[1]=1
dp[2]=2
for i in range(3,n+1):
  dp[i]=dp[i-1]+dp[i-2]
ways=dp[n]
print("Number of ways to climb stairs: ", ways)
```

2.Minimum number of coins required for target value

```python
coins=[1,2,5,10]
print("Coin values [1,2,5,10]")
target=int(input("Enter target value: "))
dp=[0]*(target+1)
for amount in range(1,target+1):
  dp[amount]=target+1 #Set it to a value greater than all possible values
  for coin in coins:
    if amount-coin >=0:
      dp[amount]=min(dp[amount], dp[amount-coin]+1)
print("Minimum coins required :", dp[target])
```

Random 1D walk

```python
import random
def random_1d_walk(step):
  current_pos=0
  left_or_right=[-1,1]
  for i in range(step):
    move=random.choice(left_or_right)
    current_pos+=move
    print("Current position:",current_pos)
  print("End position: ", current_pos)
steps=10
random_1d_walk(steps)
```
