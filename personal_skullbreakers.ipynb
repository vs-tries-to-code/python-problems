4. Contiguous subarray with maximum sum

```python
l=eval(input("Enter array: "))
max_sum=0
for i in range(len(l)):
  sub=[]
  for j in range(i, len(l)):
    sub.append(l[j])
  s=sum(sub)
  if s>max_sum:
    max_sum=s
print("Maximum sum of elements in contiguous sub arrays: ", max_sum)
```

5. Generate all permutations of a string, recursively

---



```python
def permute(s, left, right):
  if left==right:
    return (s)
  else:
    for i in range(left, right+1):
      l=list(s)
      l[left],l[i]=l[i],l[left]
      return permute("".join(s), left+1, right)

s=input("Enter string: ")
print(permute(s,0,len(s)-1))
```

9. Finding unique triplets in a list that add upto zero

```python
l=eval(input("Enter list:"))
triplets=[]
for i in range(len(l)):
  for j in range(i+1, len(l)):
    for k in range(j+1, len(l)):
      if l[i]+l[j]+l[k]==0:
        triplets.append((l[i],l[j],l[k]))
print(triplets)
```

# **Dynamic Programming**

1. Climbing stars having n steps

```python
n=int(input("Enter number of steps in stairs: "))
dp=(n+1)*[0]
dp[0]=0
dp[1]=1
dp[2]=2
for i in range(3,n+1):
  dp[i]=dp[i-1]+dp[i-2]
ways=dp[n]
print("Number of ways to climb stairs: ", ways)
```

2.Minimum number of coins required for target value

```python
coins=[1,2,5,10]
print("Coin values [1,2,5,10]")
target=int(input("Enter target value: "))
dp=[0]*(target+1)
for amount in range(1,target+1):
  dp[amount]=target+1 #Set it to a value greater than all possible values
  for coin in coins:
    if amount-coin >=0:
      dp[amount]=min(dp[amount], dp[amount-coin]+1)
print("Minimum coins required :", dp[target])
```

Distinct ways to travel from top-left to bottom right of a grid.

```python
n=int(input("Enter number of rows for square grid"))
grid=[]
for row in range(n):
  grid+=[[0]*n]
print(grid)
for i in range(n):
  for j in range(n):
    if i==0 or j==0:
      grid[i][j]=1
    else:
      grid[i][j]=grid[i][j-1]+grid[i-1][j]
print(grid)
print(grid[n-1][n-1])





```
